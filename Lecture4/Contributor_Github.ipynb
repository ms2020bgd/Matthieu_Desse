```python
#Extract all the users
def get_list_users(url):
    page = requests.get(url)
    soup = BeautifulSoup(page.text, "html.parser")
    
    table2 = soup.find('tbody').findAll('tr')
    users = []
    for i in table2:
        coco = i.find('td').a.text
        users.append(coco)
    return users
```

```python
def convert(arg):
    if "k" in arg:
        return float(re.sub("k","0",arg))*1000
    else:
        return arg
```

```python
def get_mark(user):
    #parsing step
    url = "https://github.com/"+user
    page = requests.get(url)
    soup = BeautifulSoup(page.text, "html.parser")
    
    details = soup.findAll("span", class_="Counter hide-lg hide-md hide-sm")
    repo = details[0].text.strip()
    stars = details [2].text.strip()
    #print(repo,stars)
    try:
        repo = float(convert(str(repo)))
        stars = float(convert(str(stars))) 
        mark = round(stars/repo,2)
        return mark
    except Exception:
        pass          

get_mark("andrew")
```
