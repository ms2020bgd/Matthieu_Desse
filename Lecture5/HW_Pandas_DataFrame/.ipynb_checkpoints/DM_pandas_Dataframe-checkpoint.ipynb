```python
#get country 2 digits
def getCountry(arg):
    access_token = '0a84de49aed663'
    handler = ipinfo.getHandler(access_token)
    ip_address = arg
    if checkCodeStatus(arg) != 200:
        return "N/A"
    else:
        details = handler.getDetails(ip_address)
        if "country" in details.all.keys():
            return details.all["country"]
        else:
            return "N/A"
    
    
#Check if the url exists
def checkCodeStatus(arg):
    url = "https://ipinfo.io/"+arg
    response = requests.get(url)
    return response.status_code

#66.666.666.666

#tests
print(getCountry("150.214.20.140"))
print(getCountry("666.666.666.666"))
print(getCountry("26.191.237.49"))

```

```python
def getCurrency(arg):
    code = arg
    if arg == "N/A":
        return "N/A"
    else:
        country = py.countries.get(alpha_2=code)
        print(country)
        currency = py.currencies.get(numeric=country.numeric)
        return currency
            
```

```python
def getCurrency(arg):
    url = "http://country.io/currency.json"
    response = requests.get(url).json()
    dict = response
    if arg in dict :
        if arg == "N/A":
            return "N/A"
        else:
            return dict[arg]
    else:
        return "N/A"
```

```python
df["currency"] = df["country"].apply(lambda x:getCurrency(x))
df.head(5)
```

```python
def getExchangeRate(arg):
    if arg == "N/A":
        return "N/A"
    else:
        url = "https://api.exchangerate-api.com/v4/latest/" + arg
        dict = requests.get(url).json()
        if arg in dict["rates"]:
            return dict["rates"]["EUR"]
        else: 
            return "N/A"


getExchangeRate("CHF")
```

```python
df["convertedPrice"] = df["currency"].apply(lambda x: getExchangeRate(x))
df.head(5)
```

```python
#float conversion with a miw between floats and strings

df["price"] = pd.to_numeric(df["price"],errors='coerce')
df["convertedPrice"] = pd.to_numeric(df["convertedPrice"],errors='coerce')*df["price"]
df["infos"] = df["infos"].str.lower()
df.head(20)
```

```python
#cleaning data1
uselessWords = ["contains","may","ingredients",":",",","contain",": ","and",""]
df["infos"] = df["infos"].replace(to_replace =uselessWords, value =" ") 
df.head()
```

```python
#cleaning data2
pat = r'\b(?:{})\b'.format('|'.join(uselessWords))
df['infos'] = df['infos'].str.replace(pat, '')
df["infos2"] = df['infos'].str.replace("[^\w\s]","")
df.head()
```

```python
#list of aliments
result=df["infos"].values.tolist()
result2 = [x.split(' ') for x in result]

def flatList(arg):
    flat_list = []
    for sublist in arg:
        for item in sublist:
            flat_list.append(item)
    mylist = dict.fromkeys(flat_list)
    keys=[i for i in mylist.keys()]
    #keys.remove("")
    return keys
```
